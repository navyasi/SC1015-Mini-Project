```python
#Qn: find out how differet variables predict the vote_average rating given by audience
# Basic Libraries
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('TMDB_movie_dataset_v11.csv')
df.head()
```

```python
df.info()
```

```python

def remove_columns(df, columns_to_remove):
    
    remaining_columns = [col for col in df.columns if col not in columns_to_remove]
    return df[remaining_columns]


columns_to_remove=['homepage','backdrop_path','production_companies', 'tagline','imdb_id','original_title','spoken_languages','production_countries']
new_df=remove_columns(df,columns_to_remove)

new_df.info()
```

```python
new_df = new_df[~new_df['status'].isin(['released'])]
columns_to_remove=['status']
new_df=remove_columns(new_df,columns_to_remove)
new_df=new_df.dropna()
new_df
```

```python
new_df['release_year'] = new_df['release_date'].str.extract(r'(\d{4})-\d{2}-\d{2}', expand=False)
new_df['release_year'] = new_df['release_year'].astype(int)
#calc length of chars for the title
new_df['num_chars'] = new_df['title'].str.len()

new_df
```

```python
new_df = new_df.drop(new_df[new_df['budget'] == 0].index)
new_df = new_df.drop(new_df[new_df['revenue'] == 0].index)
new_df = new_df.drop(new_df[new_df['vote_count'] == 0].index)

new_df
```

```python
new_df.to_csv('newest_dataf.csv')
```
